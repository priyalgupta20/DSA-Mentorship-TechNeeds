# **Importing Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
from sklearn.metrics import confusion_matrix

# Models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
```

# **Importing Dataset**

# **Data Preprocessing**

```python
df['Relative Humidity in %']=df.RH
df['Wind speed in km/h']=df.Ws
df['Forest Fire Occurred']=df.Classes
df=df.drop(['RH','Ws','Classes'],axis=1)
df.head()
```

```python
df=df.drop(['day','month','year','Region'],axis=1)
df.head()
```

```python
#converting categorical data to numerical form
le=LabelEncoder()
df['Forest Fire Occurred']=le.fit_transform(df['Forest Fire Occurred'])
df.head()
```

```python
#Input
X=df.drop(['Forest Fire Occurred'],axis='columns')
X.head()
```

```python
#Target
y=df['Forest Fire Occurred']
y.head()
```

```python
#Standardizing input
scaler=StandardScaler()
X=scaler.fit_transform(X)
```

```python
#Applying PCA
pca=PCA(0.95)
X=pca.fit_transform(X)

#pca gives numpy array so converting it back to df
X=pd.DataFrame(X)
X.head()
```

```python
#splitting data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

# **Model Selection and Training**

```python
#Logistic Regression

lr=LogisticRegression()
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
#Random Forest Classifier

rfc=RandomForestClassifier(random_state=1)
rfc.fit(X_train,y_train)
rfc.score(X_test,y_test)
```

```python
# Support Vector Machine Classifier

svc=SVC()
svc.fit(X_train,y_train)
svc.score(X_test,y_test)
```

```python
#Gradient Boosting Machine Classifier

gbm_clf=GradientBoostingClassifier()
gbm_clf.fit(X_train,y_train)
gbm_clf.score(X_test,y_test)
```

# **Performance Evaluation**

```python
models={
    'Logistic Regression': lr,
    'Random Forest Classifier': rfc,
    'Support Vector Classifier': svc,
    'Gradient Boosting Classifier': gbm_clf
}

scores=[]

for name, model in models.items():
    model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    scores.append({
        'Model': name,
        'Accuracy':accuracy_score(y_test,y_pred),
        'Precision':precision_score(y_test,y_pred),
        'Recall':recall_score(y_test,y_pred),
        'F1':f1_score(y_test,y_pred)
    })
```

```python
df=pd.DataFrame(scores)
df
```

# **Visualization**

```python
%matplotlib inline
plt.barh(df.Model,df.Accuracy,color='gold')
plt.xlabel('Accuracy')
plt.show()
```

```python
%matplotlib inline
plt.barh(df.Model,df.Recall,color='lightgreen')
plt.xlabel('Recall')
plt.show()
```

```python
%matplotlib inline
plt.barh(df.Model,df.f1,color='cyan')
plt.xlabel('F1 Score')
plt.show()
```
